```python
# 1 Player, Multiplayer, Online

def corregirPlayer(player):
  if 'No' in player or 'not' in player or player == '1 Player':
    return '1 Player'
  elif 'Online' in player:
    return 'Online'
  else:
    return 'Multiplayer'
```

```python
df_juegos['No_Players'] = df_juegos['No_Players'].astype(str)
```

```python
df_juegos['No_Players'] = df_juegos['No_Players'].apply(corregirPlayer)
```

```python
df_juegos['No_Players'].value_counts()
```

```python
df_juegos['Genre'].value_counts()
```

```python
df_juegos['Genre'] = df_juegos['Genre'].str.split(';')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()

encoded = pd.DataFrame(mlb.fit_transform(df_juegos['Genre']), columns=mlb.classes_, index=df_juegos.index)
df_juegos = pd.concat([df_juegos, encoded], axis=1)

df_juegos = df_juegos.drop(labels='Genre', axis=1)
```
